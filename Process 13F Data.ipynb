```python
from bs4 import BeautifulSoup as BS
import pandas as pd
import numpy as np
import requests
from yahoo_fin import stock_info
import yfinance as yf
import re
import random
import time
import PriceData
import pickle
```

```python
funds_to_stocks = pickle.load(open("funds_to_stocks.pickle","rb"))
corr_tickers = pd.read_csv('ALL_STOCKS_CORR.csv')
ticker_sector_df = pd.read_csv('ticker sector df new.csv')
ticker_sector_df = ticker_sector_df[['sector','ticker']]
cusip_tickers = pickle.load(open("cusip_tickers.pickle","rb"))
```

```python
corr_tickers = set(list(corr_tickers['ticker1']) + list(corr_tickers['ticker2']))
```

```python
ticker_sector_set = set(ticker_sector_df['ticker'].values)
```

```python
len(ticker_sector_set)
```

```python
#### Create Dict from funds to stocks
#### Key = Ticker, Val = Name
#### If Some of these tickers are in sector df
#### Join them to sector df as dont need to fetch their sector
fund_stock_tickers = {}

for fund, holdings in funds_to_stocks.items():
    
    for h in holdings:
        if h['TICKER'] not in fund_stock_tickers:
            fund_stock_tickers[h['TICKER']] = h['NAME_OF_ISSUER']
```

```python
tickers_to_merge = dict((i,v) for i,v in fund_stock_tickers.items() if i in ticker_sector_df['ticker'].values)
tickers_to_fetch_sector = dict((i,v) for i,v in fund_stock_tickers.items() if i and i not in ticker_sector_df['ticker'].values)
```

```python
len(tickers_to_fetch_sector)
```

```python
len(tickers_to_merge)
```

```python
##### merge tickers and sector df
df_tickers_to_merge = pd.DataFrame(tickers_to_merge.items(),columns=['ticker','name'])
```

```python
ticker_sector_df = ticker_sector_df.merge(df_tickers_to_merge,how='left',on='ticker')
```

```python
ticker_sector_df.to_csv("Ticker Sector Df w funds 1.csv")
```

```python
############################
```

```python
ticker_names = {"ticker":[],"sector":[]}
failed = 0
count = 0
for ticker in tickers_to_fetch_sector:
    count += 1
    print(ticker)
    try:
        sector = yf.Ticker(ticker).info['sector']
    
        ticker_names['ticker'].append(ticker)
        ticker_names['sector'].append(sector)
    except:
        print("Failed : ", ticker)
        print("Count: ", count)
        failed += 1
```

```python
#### Merge these tickers for which just fetched sector with sector df
### make dataframe and merge with fund stock tickers to add the Name of issuer
```

```python
#### ticker to name df 
fund_stock_tickers_name_df = pd.DataFrame(fund_stock_tickers.items(),columns=['ticker','name'])
```

```python
ticker_sector_df = ticker_sector_df.append(pd.DataFrame(ticker_names).merge(fund_stock_tickers_name_df,how='left',on='ticker'))
```
